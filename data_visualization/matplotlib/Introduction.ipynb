```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
 

```

```python
plt.title("Performance Comparison")
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.plot(years, runs)
```

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]

plt.plot(years, kohli,'mH--' ,label="Virat Kohli",linewidth=5)
plt.plot(years, sehwag,'cP-.', label="Virender Sehwag",linewidth=2)

plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison")
plt.legend()
plt.show()
```

```python
plt.plot(years, kohli, color='orange', linestyle='--', label="Kohli")
plt.plot(years, sehwag, color='green', linestyle='-.', label="Sehwag")
plt.plot(years, runs, color='blue', label="Tendulkar")
plt.legend()
```

```python
import matplotlib.pyplot as plt

years = [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]
kohli = [1054, 1380, 1215, 1460, 1324, 1237, 842, 930, 1100, 980]
rohit = [660, 914, 1200, 1292, 1035, 1100, 748, 820, 910, 970]
dhoni = [721, 800, 742, 890, 705, 612, 503, 620, 560, 580]

plt.plot(years, kohli, 'ro--', label='Virat Kohli')
plt.plot(years, rohit, 'm^:', label='Rohit Sharma')
plt.plot(years, dhoni, 'cP-', label='MS Dhoni')

plt.xticks(years)  # Ensure all 10 years are shown
plt.yticks(range(500, 1601, 100))  # Y-axis ticks every 100 runs

plt.title("Runs Scored (2014–2023)")
plt.xlabel("Year")
plt.ylabel("Runs")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

years = [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]

kohli = [50, 66, 89, 77, 102, 65, 88, 91, 76, 84]
rohit = [76, 58, 95, 80, 60, 72, 81, 70, 63, 67]
dhoni = [60, 65, 70, 68, 75, 62, 58, 61, 66, 59]

plt.plot(years, kohli, "ro--", label="Kohli")
plt.plot(years, rohit, "m^:", label="Rohit")
plt.plot(years, dhoni, "cP-", label="Dhoni")

# Correct way to set ticks
plt.yticks(range(50, 111, 10))  # From 50 to 110 with step of 10
plt.xticks(years)

plt.title("Runs Scored (2014–2023)")
plt.xlabel("Years")
plt.ylabel("Runs")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
