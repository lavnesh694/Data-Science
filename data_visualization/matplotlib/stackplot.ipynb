```python
import matplotlib.pyplot as plt
```

```python
label=["Studying","Cricket","Exercise","Others"]
# plt.style.use('dark_background')
plt.style.use('fast')
hours=[8,2,2,12]
plt.pie(hours,labels=label,explode=[0,0.2,0,0],startangle=360,shadow=True,wedgeprops={'edgecolor':'purple','hatch':'|','linewidth':3},textprops={'fontweight':'bold'})
plt.legend(loc="lower left")
plt.title("Student day Analysis")
```

# Now we will implement the stack plot

```python
import numpy as np

cricket=[10,9,8,7,6,5,4]
study=[2,3,4,5,6,7,8]
sleep=[9,9,9,9,3,9,9]
exercisr=[3,3,3,3,9,3,3]

days=np.arange(1,8)
label=['cricket','study','sleep','exercisr']

print(cricket,study,sleep,exercisr)
```

```python
plt.title("stack_plot")
plt.stackplot(days,cricket,study,sleep,exercisr,labels=label)
plt.legend()
plt.grid(True)
# plt.xlabel(range(1,25,1))
```
