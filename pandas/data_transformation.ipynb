```python
data={
    'NAME':["Lavnesh","Deepak","Anant"],
    'MATH':[100,90,80],
    'SCIENCE':[90,82,89],
    'ENGLISH':[88,85,94]

}

data
```

```python
df=pd.DataFrame(data)
df['total']=df.eval("MATH + SCIENCE + ENGLISH")
```

```python
df
```

```python
df1=df.melt(id_vars=["NAME"],value_vars=["MATH","SCIENCE","ENGLISH"],var_name="Subject",value_name="Score")

df1['grade']=df1['Score'].apply(lambda x: "GOOD MARKS" if x>=90 else "BURE NO")

df1
```
