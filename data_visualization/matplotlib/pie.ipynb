```python
# Data

import matplotlib.pyplot as plt
plt.style.use("") 
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
explode = [0.4,0,0,0]  # Only highlight Sachin's slice


runs = [18000, 8000, 12000, 9500]

plt.title("Career Runs of Indian Batsmen")
# plt.pie(runs, labels=labels)
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']


plt.pie(
    runs,
    labels=labels,
    shadow=True,
    colors=colors,
    explode=explode,
    startangle=540,
      # autopct='%1.1f%%',
    
    wedgeprops={'edgecolor': 'white', 'linewidth': 2, 'linestyle': '-','hatch':'.','snap':True}
)
plt.show()
```
