```python
import pandas as pd
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
data=pd.read_excel('iris.xlsx')
data.head()
```

```python
x=data.iloc[:,:-1]
```

```python
y=data.iloc[:,-1]
```

```python
y
```

```python
model =DecisionTreeClassifier()
```

```python
model.fit(x,y)
```

```python
prediction=model.predict([[6.4,	7.4	,8.4,	9.4]])
```

```python
prediction
```

```python
train=pd.read_excel('irist.xlsx')
```

```python
xt=train.iloc[:,:-1]
```

```python
xt
```

```python
prediction=model.predict(xt)
```

```python
prediction
```

```python
y_test=train.iloc[:,-1]
```

```python
actual=y_test.values
actual
```

```python
count=0
for i in range(0,len(actual)):
    if actual[i]==prediction[i]:
        count=count+1

total_accuracy=round(count*100/len(actual),2)
total_accuracy
```
